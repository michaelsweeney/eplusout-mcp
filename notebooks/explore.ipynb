```python

dfconcat = pd.concat([pd.DataFrame(x) for x in resultlist])
dfconcat.set_index(['dt', 'KeyValue', 'Name'])['Value']
dfconcat = dfconcat[['dt', 'KeyValue', 'Name', 'Value']]
dfconcat['htg_or_cool'] = np.where(
    dfconcat['Name'] == "Zone Water to Air Heat Pump Total Heating Rate", "Heating", "Cooling"
)

dfconcat  = dfconcat[['dt', 'KeyValue', 'htg_or_cool', 'Value']]

dfgrp = dfconcat.groupby(['dt', 'htg_or_cool'])['Value'].sum().reset_index()
dfgrp = dfgrp.set_index(['dt', 'htg_or_cool']).unstack('htg_or_cool')

dfgrp.columns = [x[1] for x in dfgrp.columns]
dfgrp['net'] = dfgrp['Cooling'] - dfgrp['Heating']

dfgrp['dayofyear'] = dfgrp.index.dayofyear
dfgrp['date'] = dfgrp.index

dfgrp['overlap_flag'] = ((dfgrp['Heating'] > 0) & (dfgrp['Cooling'] > 0)).astype(int)

## get daily aggregation
daily = dfgrp.groupby('date').agg(
    heating_sum=('Heating', 'sum'),
    cooling_sum=('Cooling', 'sum'),
    heating_hours=('Heating', lambda x: (x > 0).sum()),
    cooling_hours=('Cooling', lambda x: (x > 0).sum()),
    overlap_hours=('overlap_flag', 'sum')
).reset_index()



# # # Filter candidate days
candidates = daily[(daily['heating_sum'] > 0) & (daily['cooling_sum'] > 0)]
candidates['balance'] = candidates['cooling_sum'] / candidates['heating_sum']

# # Rank by balance close to 1 and overlap presence
candidates['balance_score'] = (candidates['balance'] - 1).abs()
best_days = candidates.sort_values(['balance_score', 'overlap_hours']).head(5)

# get date of highest balance_score
selected_date = best_days.loc[best_days['balance_score'].idxmax()]['date']


selected_date = best_days.iloc[1]['date']

selected = dfgrp.loc[dfgrp.dayofyear == selected_date.dayofyear]
px.bar(selected[['Cooling', 'Heating', 'net']], barmode='group')
```
